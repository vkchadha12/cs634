```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import pickle
from pickle import dump
from pickle import load

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

scaler = MinMaxScaler()

np.random.seed(42)


st.markdown("", unsafe_allow_html=True)

st.markdown("House Price Prediction in Ames,Iowa", unsafe_allow_html=True)

st.markdown("A multi-step process is used to estimate the range of house prices based on your selection. The modeling process is done using the data found on Kaggle(link at left bottom corner of page)", unsafe_allow_html=True)


name_list = ['MSSubClass',
 'OverallQual',
 'GrLivArea',
 'TotalBsmtSF',
 'GarageCars',
 'BsmtFinSF1',
 'FstFlrSF',
 'SndFlrSF',
 'YearRemodAdd',
 'YearBuilt',
 'LotArea',
 'Fireplaces',
 'FullBath',
 'HalfBath'
 ]

name_list_train = ['MSSubClass',
 'OverallQual',
 'GrLivArea',
 'TotalBsmtSF',
 'GarageCars',
 'BsmtFinSF1',
 '1stFlrSF',
 '2ndFlrSF',
 'YearRemodAdd',
 'YearBuilt',
 'LotArea',
 'Fireplaces',
 'FullBath',
 'HalfBath'
]

data = pd.read_csv('train.csv')
data = data[name_list_train].values

scaler.fit(data)

description_list = [
 'What is the building class?',
 'What is the Overall material and finish quality?',
 'What is the Above grade (ground) living area square feet?',
 'What is the Unfinished square feet of basement area?',
 'What is Size of garage in car capacity',
 'What is the Type 1 finished square feet',
 'What is the First Floor square feet?',
 'What is the Second floor square feet?',
 'In which year was it remodelled?',
 'In which year was it Built?',
 'What is the LotArea',
 'What is the number of fireplaces?',
 'What is the number of FullBath?',
 'What is the number of HalfBath?',

 ]

min_list = [
20,
0,
334,
1950.0,
0,
0.0,
334.0,
0.0,
334.0,
1872,
1300,
0.0,
0.0,
0.0,
]


max_list =[
190,
10.0,
5642,
2010.0,
4,
5644,
5642.0,
2065.0,
5642.0,
2010,
215245,
3.0,
3.0,
2.0]

count = 0

with st.sidebar:

    for i in range(len(name_list)):
        variable_name = name_list[i]
        globals()[variable_name] = st.slider(description_list[i] ,min_value=int(min_list[i]), max_value =int(max_list[i]),step=1)

userinput_data_df = {
 'MSSubClass': [MSSubClass],
 'OverallQual': [OverallQual],
 'GrLivArea':[GrLivArea],
 'TotalBsmtSF': [TotalBsmtSF],
 'GarageCars': [GarageCars],
 'BsmtFinSF1': [BsmtFinSF1],
 '1stFlrSF': [FstFlrSF],
 '2ndFlrSF': [SndFlrSF],
 'YearRemodAdd': [YearRemodAdd],
 'YearBuilt': [YearBuilt],
 'LotArea': [LotArea],
 'Fireplaces': [Fireplaces],
 'FullBath' :[FullBath],
 'HalfBath' : [HalfBath]}


negloglik = lambda y, p_y: -p_y.log_prob(y) # note this

rf_model = pickle.load(open('rf_model.pk1', 'rb'))
dc_model = pickle.load(open('dc_model.pk1', 'rb'))
lgbm_model = pickle.load(open('lgbm_model.pk1', 'rb'))

userinput_data_df = pd.DataFrame.from_dict(userinput_data_df)
st.dataframe(userinput_data_df)
rf_model.predict(userinput_data_df)
userinput_data_df_normal = scaler.transform(userinput_data_df)
rf_yhat = rf_model.predict(userinput_data_df_normal)
dc_yhat = dc_model.predict(userinput_data_df_normal)
lgbm_yhat = lgbm_model.predict(userinput_data_df_normal)


rf_lower_number = "{:,.2f}".format(int(rf_yhat.mean() - 1.95 * rf_yhat.std()))
rf_higher_number = "{:,.2f}".format(int(rf_yhat.mean() + 1.95 * rf_yhat.std()))
dc_lower_number = "{:,.2f}".format(int(dc_yhat.mean() - 1.95 * dc_yhat.std()))
dc_higher_number = "{:,.2f}".format(int(dc_yhat.mean() + 1.95 * dc_yhat.std()))
lgbm_lower_number = "{:,.2f}".format(int(lgbm_yhat.mean() - 1.95 * lgbm_yhat.std()))
lgbm_higher_number = "{:,.2f}".format(int(lgbm_yhat.mean() + 1.95 * lgbm_yhat.std()))

col1, col2, col3 , col4, col5 = st.columns(5)

with col1:
    pass
with col2:
    pass
with col4:
    pass
with col5:
    pass
with col3 :
    center_button = st.button('Calculate range of house price')



if center_button:
    import time
    #my_bar = st.progress(0)
    with st.spinner('Calculating....'):
        time.sleep(2)

    st.markdown("The price is Calculated using RandomForestRegressor & DecisionTreeRegressor & LGBMRegressor:", unsafe_allow_html=True)
    col1, col2 = st.columns([3, 3])


    col1, col2, col3 = st.columns(3)



    with col1:
        st.write("")

    with col2:
        st.subheader("RandomForestRegressor Predicton $"+ str(rf_higher_number))
        st.subheader("       AND ")

        st.subheader("DecisionTreeRegressor Predicton $"+str(dc_higher_number))
        st.subheader("       AND ")
        st.subheader("LGBMRegressor $"+str(lgbm_higher_number))




    with col3:
        st.write("")
```
